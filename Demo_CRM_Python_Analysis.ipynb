**# 1. ANALYSIS: Age & Gender Segmentation for Campaign Targeting**

```python
# Group by gender and calculate age stats

age_gender = df.groupby("gender")["age"].agg(["count", "mean", "min", "max"]).reset_index()
print(age_gender)
```

**# 2. VISUAL: Role Distribution**

```python
# Group contact data by 'role' and count how many contacts belong to each role
# Useful to identify which roles (e.g., 'admin', 'manager', 'user') are most common
# Helps prioritize decision-makers for upsell or outreach efforts

role_counts = df["role"].value_counts().reset_index()
role_counts.columns = ["role", "count"]  # Rename columns for clarity

# Print the top roles

print("🔍 Contact distribution by role:")
print(role_counts.head())
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=role_counts, x="count", y="role", hue="role", palette="coolwarm", legend=False)
plt.title("Contacts by Role (Potential Decision Makers)")
plt.xlabel("Number of Contacts")
plt.ylabel("Role")
plt.tight_layout()
plt.show()
```

**# 3. Email Domain Segmentation**

```python
# Search for B2B domains (e.g. @company.com) vs. free (@gmail.com, @yahoo.com)
# Target decision-makers using company domains
# Prioritize contacts with business emails — they’re more likely to represent companies.

df["is_business_email"] = df["email_domain"].apply(lambda x: "gmail" not in x and "yahoo" not in x and "outlook" not in x)
print(df[["email", "email_domain", "is_business_email"]])
```

**# 4. Age-Based Opportunity Filtering**

```python
# This analysis segments contacts by age groups commonly used in CRM strategy:
# - 30+ = Likely buying decision-makers
# - 20–30 = Early adopters and digital-first buyers
# - <20 = Long-term potential or high lifetime value (LTV) prospects
# This helps tailor messaging, offer timing, and upsell priorities by age cohort.


df["age_segment"] = pd.cut(df["age"], bins=[0, 25, 35, 60], labels=["Gen Z", "Millennial", "Gen X+"])
age_segment_counts = df["age_segment"].value_counts()
print(age_segment_counts)
```

**# 4. Educational Background Clustering**

```python
# Group contacts by university to spot shared academic backgrounds.
# Useful for alumni-targeted outreach or understanding education level clustering.

university_counts = df["university"].value_counts().reset_index()
university_counts.columns = ["university", "contact_count"]

print("🎓 Contacts per University:")
print(university_counts)

# Optional visual
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6, 4))

sns.barplot(
    data=university_counts,
    x="contact_count",
    y="university",
    hue="university",          # Use same as y-axis
    palette="Blues_d",
    legend=False               # Hide duplicate legend
)

plt.title("University Clustering of Contacts")
plt.xlabel("Number of Contacts")
plt.ylabel("University")
plt.tight_layout()
plt.show()
```

**# 5. Digital/Technical Environment Analysis**



```python
# Extract browser/OS info from userAgent string
# Helps identify user's tech environment for segmentation or support targeting

df["browser_info"] = df["userAgent"].str.extract(r'\((.*?)\)')

# Display parsed userAgent info (e.g., OS, device)
print("🔍 Parsed browser/device environment from userAgent:")
print(df[["userAgent", "browser_info"]])
```

```python
# Count unique IP addresses
# Useful for detecting unique devices or sessions per contact

unique_ips = df["ip"].nunique()
print(f"📍 Number of unique IP addresses: {unique_ips}")
```

```python
# Count unique MAC addresses
# Can act as a unique identifier for device-based tracking

unique_macs = df["macAddress"].nunique()
print(f"📱 Number of unique MAC addresses: {unique_macs}")
```

```python
# Combine IP, MAC, and browser info to create a basic device fingerprint
# Helps simulate digital identity for account tracking or fraud detection

df["device_fingerprint"] = df["ip"] + " | " + df["macAddress"] + " | " + df["browser_info"].fillna("Unknown")

# Show sample device fingerprints
print("🧬 Device fingerprint per contact:")
print(df[["first_name", "device_fingerprint"]])
```

```python
# Count how many times each unique browser/OS string appears
# Note that 'browser_info' was extracted from userAgent earlier (e.g., Windows NT, Mac OS X)

browser_counts = df["browser_info"].value_counts().reset_index()
browser_counts.columns = ["browser_os", "count"]  # Rename columns for clarity

# 📊 Create a horizontal bar plot to visualize browser/OS distribution
plt.figure(figsize=(8, 4))  # Set the chart size

sns.barplot(
    data=browser_counts,     # Use the counted values
    x="count",               # Bar length = number of contacts
    y="browser_os",          # Each bar represents a browser/OS string
    hue="browser_os",        # Assign hue to match the y-axis variable
    palette="coolwarm",      # Use a colorful palette for clarity
    legend=False             # Hide redundant legend
)

# 🏷️ Add titles and labels to explain the chart
plt.title("Distribution of Browser/OS Environments (userAgent)")
plt.xlabel("Number of Contacts")
plt.ylabel("Browser / OS Info")

# 💡 Use tight layout so labels don't get cut off
plt.tight_layout()

# 📈 Show the plot in the notebook
plt.show()
```
